# Movie Reccommendations 

```python
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
```

### Loading Movie DataSet

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2))

ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index='user_id',columns='title',values='rating')
movieRatings.head()
```

```python
print type(movieRatings)
print movieRatings.shape
```

### Similar Movies using Correlation 



```python
movie = movieRatings['Star Wars (1977)']
movie.head(15)
#type(movie)
```

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

```python
def simMovies (movie):
    similarMovies = movieRatings.corrwith(movie)
    similarMovies = similarMovies.dropna()
    df = pd.DataFrame(similarMovies, columns=['similarity'])
    df = df.reset_index(level=['Title'])
    return df,similarMovies

df,similarMovies = simMovies(starWarsRatings)
df.head(10)
```

```python
similarMovies = similarMovies.sort_values(ascending=False)
similarMovies.head(10)
#type(similarMovies)
```

### Movie Rating stats

#### Count, Mean Ratings

```python
movieStats = ratings.groupby('title').agg({'rating':[np.size, np.mean]})
movieStats.head(10)
```

### Filter movie Sets based on number of Ratings, use ratings with count > 100

```python
popularMovies = movieStats[movieStats['rating']['size'] >= 200]
#popularMovies.head(10)
popularMovies.sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = popularMovies.join(pd.DataFrame(similarMovies, columns=['similarity']))
df.head(15)
```

```python
similar_df = df.sort_values(['similarity'], ascending=False)[:15]
similar_df
```

## Item Based Collaborative Filtering 

```python
userRatings = ratings.pivot_table(index='user_id',columns='title',values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.tail()
```

```python
myRatings = userRatings.loc[1].dropna()
#myRatings = myRatings.head(5)
myRatings.head(10)
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    #print "Adding sims for " + myRatings.index[i] + "..."
    
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()

    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    #print sims
    
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at the results so far:
print "sorting..."
simCandidates.sort_values(inplace = True, ascending = False)
print simCandidates.head(10)
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(15)
```
